```python
import cv2
import tensorflow as tf
from keras.preprocessing import image
from PIL import Image
import numpy as np
```

```python
model = tf.keras.models.load_model(r'BrainTumor_model')
```

```python
def names(number):
    if number==0:
        return 'No, Its not a tumor'
    else:
        return 'Yes ! Its a Tumor'
```

```python
# model = tf.keras.models.load_model(r'BrainTumor_model')

from matplotlib.pyplot import imshow
img = Image.open(r"C:\Users\meabh\Deep Learning\Brain_tumor_detector\n.jpg")
x = np.array(img.resize((150,150)))
x = x.reshape(1,150,150,3)
res = model.predict_on_batch(x)
classification = np.where(res == np.amax(res))[1][0]
imshow(img)
print(str(names(classification)))
```
